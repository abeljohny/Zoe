```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# responses from 5 participants for 16 questions
data = {
    "Question 01": [4, 3, 3, 4, 4],
    "Question 02": [2, 2, 3, 2, 2],
    "Question 03": [4, 4, 4, 4, 3],
    "Question 04": [1, 1, 1, 1, 1],
    "Question 05": [5, 4, 4, 5, 5],
    "Question 06": [1, 2, 1, 1, 1],
    "Question 07": [4, 4, 3, 4, 4],
    "Question 08": [1, 2, 2, 2, 2],
    "Question 09": [4, 3, 3, 3, 4],
    "Question 10": [1, 2, 2, 2, 2],
    "Question 11": [4, 4, 4, 4, 4],
    "Question 12": [1, 2, 1, 1, 1],
    "Question 13": [5, 3, 4, 5, 4],
    "Question 14": [1, 2, 1, 1, 1],
    "Question 15": [5, 3, 3, 4, 4],
    "Question 16": [1, 2, 2, 1, 1],
}
df = pd.DataFrame(data)
# Convert DataFrame to long format for easier plotting
df_long = df.melt(var_name="Question", value_name="Response")
# Create a pivot table for the heatmap
response_pivot = df_long.pivot_table(
    index="Question", columns="Response", aggfunc=len, fill_value=0
)
# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(response_pivot, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Likert Scale Questionnaire Responses (5 Participants)")
plt.ylabel("Question")
plt.xlabel("Response")
plt.show()
```
